```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

```python

table=pd.read_csv('/content/drive/MyDrive/Dream Team Projects/intel_dt.csv')
table

```

```python
table=table.drop('Unnamed: 0',axis=1,inplace=False)
table
```

```python
bins1=np.linspace(min(table['Lat']),max(table['Lat']),9)
bins2=np.linspace(table['Long'].min(),table['Long'].max(),9)
latname=['North-Madras','South-Madras','Tambaram-Area','Tam-Chen-Area','Chengalpettu-Area','Mama-Mela-Region','Madhuranthakam-Area','Mel-Acharapakkam']
longname=['North-Madras','South-Madras','Tambaram-Area','Tam-Chen-Area','Chengalpettu-Area','Mama-Mela-Region','Madhuranthakam-Area','Mel-Acharapakkam']

table['Latitude-Region']=pd.cut(table['Lat'],bins1,labels=latname,include_lowest=True)
table['Longitude-Region']=pd.cut(table['Long'],bins2,labels=longname,include_lowest=True)
table
```

```python
table=table.drop('Latitude-Region', axis=1,inplace=False)
```

```python
table.rename(columns={'Longitude-Region':'Region'},inplace=True)
table
```

```python
table.to_csv('/content/drive/MyDrive/Dream Team Projects/intel_dt.csv')
```
