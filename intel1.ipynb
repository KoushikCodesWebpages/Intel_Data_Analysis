```python
datename=dfor['Date']
dateout=dfor["Date"]
```

```python
daylist=[]

def get_day_of_week(date_str):
    date_obj = datetime.datetime.strptime(date_str, '%Y-%m-%d')
    day_of_week = date_obj.weekday()
    day = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    return day[day_of_week]
for date_str in dateout:
    day_of_week = get_day_of_week(date_str)
    daylist.append(day_of_week)


```

```python
days=pd.DataFrame(daylist)
days['Day']=pd.DataFrame(daylist)
days.value_counts()
days.drop(0,axis=1,inplace=True)
days
```

```python
tab=pd.concat([df,days], axis=1)
tab
```

```python

tab.to_csv('/content/drive/MyDrive/Dream Team Projects/intel_dt.csv')
```

```python
grpb=tab[['Alert','Time-Hour','Lat']]
group=grpb.groupby(['Alert','Time-Hour'],as_index=False).min()
group2=grpb.groupby(['Alert','Time-Hour'],as_index=False).max()
print(group)
print(group2)
```

```python
grp_pivot=group.pivot(index='Alert',columns='Time-Hour')
print(grp_pivot)
grp_pivot2=group2.pivot(index='Alert',columns='Time-Hour')
print(grp_pivot2)
```

```python
tab['Lat'].describe()
start=tab['Lat'].min()
end=tab['Lat'].max()
linlat=np.linspace(start,end,9)
print(linlat)
lats=pd.DataFrame(linlat)
lats=lats.rename(columns={0:'Latitude'})
print(lats)
```

```python
start=tab['Long'].min()
end=tab['Long'].max()
linlong=np.linspace(start,end,9)
linlong
longs=pd.DataFrame(linlong)
longs=longs.rename(columns={0:'Longitude'})
print(longs)
```

```python
latlong=pd.concat([lats,longs],axis=1)
latlong
```
