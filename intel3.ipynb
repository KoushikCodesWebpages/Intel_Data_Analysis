```python
import numpy as np
import scipy as sp
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
table=pd.read_csv('/content/drive/MyDrive/Dream Team Projects/intel_dt.csv')
```

```python
table
```

```python
table=table.drop('Unnamed: 0',axis=1,inplace=False)
table
```

```python
grp=table[['Region','Day','fcw']]
grp=grp.groupby(['Region','Day'],as_index=False).mean()
grp_pivot=grp.pivot(index='Region',columns='Day')
grp_pivot
```

```python
grp1=table[['Region','Day','pcw']]
grp1=grp1.groupby(['Region','Day'],as_index=False).sum()
grp_pivot1=grp1.pivot(index='Region',columns='Day')
grp_pivot1
```

```python
grp2=table[['Region','Day','ldw']]
grp2=grp2.groupby(['Region','Day'],as_index=False).sum()
grp_pivot2=grp2.pivot(index='Region',columns='Day')
grp_pivot2
```

```python
grp3=table[['Region','Day','hmw']]
grp3=grp3.groupby(['Region','Day'],as_index=False).sum()
grp_pivot3=grp3.pivot(index='Region',columns='Day')
grp_pivot3
```
