```python
dftim=df['Time'].str.split(':', expand=True)
dftim[['Time-Hour','Time-Min','Time-Sec']]=df['Time'].str.split(':', expand=True)
dftim
```

```python
dftimes=dftim[['Time-Hour','Time-Min','Time-Sec']]
```

```python
df.drop('Time',axis=1, inplace=True)
```

```python
table=pd.concat([df, dftimes], axis=1)
table
```

```python
dfdate=table['Date'].str.split("-",expand=True)
dfdate[['Date-Year','Date-Month','Date-Day']]=table['Date'].str.split("-",expand=True)
dfdate=dfdate[['Date-Year','Date-Month','Date-Day']]
dfdate

```

```python
table.drop('Date',axis=1,inplace=True)
table=pd.concat([table,dfdate],axis=1)
```
